```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
print(os.path.abspath('column_info.txt'))


pd.set_option('display.max_columns', None)   
pd.set_option('display.width', None)        
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)

import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
data = pd.read_csv(r'C:\Users\91701\OneDrive\Documents\PulseGuard AI – Intelligent Blood Pressure Prediction System hackthon 2026\patient_data.csv')
print(data.head())
print('--------//--------')
print(data.tail())
```

```python
for col in data.columns:
    print(f"Column: {col}")
    print(data[col].unique())
    print("Number of unique values:", data[col].nunique())
    print("-----------")
```

```python
# Remove leading/trailing spaces
data = data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Convert everything to lowercase (very important)
data = data.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

# Normalize dash formatting
for col in ['Age', 'Systolic', 'Diastolic']:
    if col in data.columns:
        data[col] = data[col].str.replace('-', ' - ')
        data[col] = data[col].str.replace('  ', ' ')
```

```python
data['Stages'] = data['Stages'].replace({'hypertensive crisi': 'hypertensive crisis'})
data['Stages'] = data['Stages'].str.rstrip('.')
```

```python
data.columns
```

```python
data.rename(columns={'C':'Gender'}, inplace=True)

for col in data.columns:
    print(f"Column: {col}")
    print(data[col].unique())
    print("Number of unique values:", data[col].nunique())
    print("-----------")
```

```python
with open('column_info.txt', 'w') as f:
    for col in data.columns:
        f.write(f"Column: {col}\n")
        unique_vals = data[col].unique().tolist()
        for val in unique_vals:
            f.write(f"  - {val}\n")
        f.write(f"Number of unique values: {len(unique_vals)}\n")
        f.write("-----------\n")

print("Done! Check column_info.txt")
```

```python
data.isnull().sum()
```

```python
nominal_features=['Gender', 'History', 'Patient', 'TakeMedication', 'BreathShortness', 'VisualChanges', 'NoseBleeding', 'ControlledDiet']
ordinal_features=[f for f in data.columns if f not in nominal_features]
ordinal_features.remove('Stages')
print (nominal_features)
print(ordinal_features)
```

```python

# Encode Nominal Features


for col in nominal_features:
    
    if set(data[col].unique()) == {'yes', 'no'}:
        data[col] = data[col].map({'no': 0, 'yes': 1})
    
    elif col == 'Gender':
        data[col] = data[col].map({'male': 0, 'female': 1})


# Encode Ordinal Features


data['Age'] = data['Age'].map({
    '18 - 34': 1,
    '35 - 50': 2,
    '51 - 64': 3,
    '65+': 4
})

data['Severity'] = data['Severity'].replace({
    'mild': 0,
    'moderate': 1,
    'sever': 2
})

data['Whendiagnoused'] = data['Whendiagnoused'].map({
    '<1 year': 1,
    '1 - 5 years': 2,
    '>5 years': 3
})

data['Systolic'] = data['Systolic'].map({
    '100+': 0,
    '111 - 120': 1,
    '121 - 130': 2,
    '130+': 3
})


data['Diastolic'] = data['Diastolic'].map({
    '70 - 80': 0,
    '81 - 90': 1,
    '91 - 100': 2,
    '100+': 3,
    '130+': 4
})



# Encode Target

data['Stages'] = data['Stages'].map({
    'normal': 0,
    'hypertension (stage-1)': 1,
    'hypertension (stage-2)': 2,
    'hypertensive crisis': 3
})
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[ordinal_features]= scaler.fit_transform(data[ordinal_features])
```

```python
print(data.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
x= data.drop('Stages', axis=1)
y=data['Stages']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy={}
```

```python
#Logistic Regression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
print("Logistic Regression:\n ")
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f} %")
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
accuracy={'Logistic Regression' : accuracy_score(y_test, y_pred) }
```

```python
#Decision Tree

decisionTree=DecisionTreeClassifier()
decisionTree.fit(x_train, y_train)
y_pred = decisionTree.predict(x_test)
print("Decision Tree:\n")
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f} %")
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
accuracy['Decision Tree']=accuracy_score(y_test, y_pred)
```

```python
#Random Forest

randomforest=RandomForestClassifier()
randomforest.fit(x_train, y_train)
y_pred = randomforest.predict(x_test)
print("Random Forest:\n")
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f} %")
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
accuracy ['Random Forest' ]=accuracy_score(y_test, y_pred)
```

```python
#Support Vector Machine

svm=SVC()
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
print("Support Vector Machine:\n")
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f} %")
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
accuracy['SVM' ]=accuracy_score(y_test, y_pred)

```

```python
#k nearest neighbors - KNN
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print("K Nearest Neighbors:\n")
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f} %")
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
accuracy['KNN']=accuracy_score(y_test, y_pred)
```

```python
#Ridge Classifier 

RC=RidgeClassifier()
RC.fit(x_train, y_train)
y_pred = RC.predict(x_test)
print("Ridge Classifier:\n")
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f} %")
print("Classification Report: \n", classification_report(y_test, y_pred))   
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
accuracy['Ridge Classifier']=accuracy_score(y_test, y_pred)

```

```python
#Gaussian Naive Bayes

GNB=GaussianNB()
GNB.fit(x_train, y_train)
y_pred = GNB.predict(x_test)
print("Gaussian Naive Bayes:\n")
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f} %")
print("Classification Report: \n", classification_report(y_test, y_pred))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
accuracy['Gaussian Naive Bayes']=accuracy_score(y_test, y_pred)
```
